# Training on the data

* num_objects : 분류할 class의 숫자
* num_experiments : epochs
* enhance_data (optional) : 더 나은 성능을 위해 훈련 이미지의 수정 된 사본을 생성하도록 하려는 지 여부를 나타냅니다.
* batch_size : 네트워크가 하나에서 처리 할 이미지의 수를 나타냅니다.
* show_network_summary : 이는 네트워크가 콘솔에 훈련 네트워크의 구조를 표시해야하는지 여부를 나타냅니다.
* save_from_model : val_Accuracy가 더 높은 모델만 저장됨
* continue_from_model : 절대경로로 입력해서 해당 위치에 있는 모델을 불러옴

```
from imageai.Prediction.Custom import ModelTraining
import os

execution_path = os.getcwd()

DATASET_DIR = os.path.join(execution_path, "beverage")

model_trainer = ModelTraining()
model_trainer.setModelTypeAsDenseNet()
model_trainer.setDataDirectory(DATASET_DIR)
model_trainer.trainModel(num_objects=10, num_experiments=100, enhance_data=True, batch_size=32, show_network_summary=True, save_full_model=True, continue_from_model="/content/drive/My Drive/KISTI_TEAM/resize_image/model_ex-018_acc-0.879167.h5")
```

# Predict image


```
from imageai.Prediction.Custom import CustomImagePrediction
import os

execution_path = os.getcwd()

predictor = CustomImagePrediction()
predictor.setModelPath(model_path=os.path.join(execution_path, "model_ex-022_acc-0.970833.h5"))
predictor.setJsonPath(model_json=os.path.join(execution_path, "model_class.json"))
predictor.loadFullModel(num_objects=10)

results, probabilities = predictor.predictImage(image_input=os.path.join(execution_path, "images2.jpg"), result_count=5)
for i in range(len(results)):
  print(results[i],":",probabilities[i])
```
