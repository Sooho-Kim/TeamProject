```
import pandas as pd

data = pd.read_csv('log.csv', delimiter=';')
data
```

```
import matplotlib.pyplot as plt
f,ax = plt.subplots(1,2, figsize=(18,8))

ax[0].set_title('Model accuracy')
ax[0].plot(data['epoch'], data['accuracy'])
ax[0].plot(data['epoch'], data['val_accuracy'])
ax[0].legend(['acc','val_acc'])



ax[1].set_title('loss')
ax[1].plot(data['epoch'], data['loss'])
ax[1].plot(data['epoch'], data['val_loss'])
ax[1].legend(['loss','val_loss'])
plt.show()
```
